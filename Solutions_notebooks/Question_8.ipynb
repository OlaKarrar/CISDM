# Question 8

Generate points $(X, Y)$, where $Y = f(X) + \epsilon$ with given $f$, $X \sim U(-5, 5)$, and independent error $\epsilon \sim \mathcal{N}(0, \sigma^2)$. Try out kernel regression method to recover $f$. Try different settings and analyze your error. Use functions $f(x) = x$, $f(x) = x^3$, and $f(x) = \sin(x)$.

## Solution:

Now we will compare the Mean Squared Error (MSE) with different Sigma and Kernel Width values.

1. $F(x) = x$

Case 1: Lower Sigma $(0.1)$ and Kernel Width $(0.1, 1.0$ and $2.0)$

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.kernel_ridge import KernelRidge

np.random.seed(42)

# Function for kernel regression
def kernel_regression(X, Y, kernel_width=1.0):
    model = KernelRidge(kernel='rbf', gamma=1.0 / (2 * kernel_width**2))
    model.fit(X.reshape(-1, 1), Y)
    return model.predict(X.reshape(-1, 1))

def generate_data_x(sigma, size=100):
    X = np.random.uniform(-5, 5, size)
    epsilon = np.random.normal(0, sigma, size)
    Y = X + epsilon #for F(x) = x^3, Y = X**3 + epsilon and for F(x) = Sin (x), Y = np.sin(X) + epsilon
    return X, Y

sigmas_x = [0.1]
kernel_widths_x = [0.1, 1.0, 2.0]

for sigma_x in sigmas_x:
    for kernel_width_x in kernel_widths_x:

        X_x, Y_x = generate_data_x(sigma_x)
        X_x = X_x.reshape(-1, 1)

        model = KernelRidge(alpha=1.0, kernel='rbf', gamma=1.0/kernel_width_x)
        model.fit(X_x, Y_x)

        X_trend = np.linspace(X_x.min(), X_x.max(), 100).reshape(-1, 1)

        Y_trend = model.predict(X_trend)

        plt.scatter(X_x, Y_x, color='lightblue', label=f'Data points (Sigma={sigma_x}, Kernel Width={kernel_width_x})')

        plt.plot(X_trend, Y_trend, color='blue', label='Kernel Regression Trend Line', linestyle='-')

        mse_x = mean_squared_error(Y_x, model.predict(X_x))
        print(f'Mean Squared Error for f(x) = x (Sigma={sigma_x}, Kernel Width={kernel_width_x}): {mse_x:.4f}')

        plt.title(f'Kernel Regression for f(x) = x (Sigma={sigma_x}, Kernel Width={kernel_width_x})')
        plt.xlabel('X')
        plt.ylabel('Y')
        plt.legend()
        plt.show()
```

According to the above details, the accuracy of kernel regression estimates is influenced by the values of Sigma and Kernel Width.

**1. Sigma**

a. Higher Sigma: This may result in a more scattered dataset with larger errors. Kernel regression might struggle to capture the underlying pattern when there is significant noise in the data.

b. Lower Sigma: A lower sigma implies less noise in the data, making it easier for kernel regression to capture the true underlying function.

**2. Kernel Width**

a. Higher Kernel Width: A larger kernel width makes the regression smoother and may lead to over smoothing or under fitting. It may fail to capture local variations in the data, resulting in a biased estimate.

b. Lower Kernel Width: A smaller kernel width allows the regression to be more sensitive to local variations, which can be beneficial when the underlying function changes rapidly. However, too low a value might lead to overfitting, capturing noise in the data.
